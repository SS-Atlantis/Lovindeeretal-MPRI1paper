```python
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
# Load data (example)
df = pd.read_csv("biomass_2007.csv", parse_dates=["Date"])

date_col = "Date"
biomass_cols = [c for c in df.columns if c != date_col]

n_cols = 4
n_rows = math.ceil(len(biomass_cols) / n_cols)

fig, axes = plt.subplots(
    n_rows, n_cols,
    figsize=(20, n_rows * 2),
    sharex=True
)

axes = axes.flatten()

for i, col in enumerate(biomass_cols):
    axes[i].plot(df[date_col], df[col], linewidth=1)
    axes[i].set_title(col, fontsize=14)
    axes[i].grid(True, alpha=0.3)

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

#fig.suptitle("Biomass Time Series", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
fig.savefig('Figures/FigS2_control_biomass.png', dpi=300, bbox_inches='tight')
```
